```python
#Exercise 2
import numpy as np
import pandas as pd

url = "https://assets.01-edu.org/ai-branch/piscine-ai/household_power_consumption.txt"
data = pd.read_csv(url, sep=';', low_memory=False)

# 1 Delete the columns Time, Sub_metering_2 and Sub_metering_3
data.drop(columns=['Time', 'Sub_metering_2', 'Sub_metering_3'], inplace=True)


#2- Set Date as index
print("\n2\n")
# Convertir la colonne Date en datetime
data['Date'] = pd.to_datetime(data['Date'], format='%d/%m/%Y')
# Définir la colonne Date comme index
data.set_index('Date', inplace=True)
print(data.head().index)

#3- Create a function that takes as input the DataFrame with the data set and returns a DataFrame with updated types:
print("\n3\n")
def update_types(df):
    # Convert relevant columns to float64
    float_columns = [
        'Global_active_power',
        'Global_reactive_power',
        'Voltage',
        'Global_intensity',
        'Sub_metering_1'
    ]
    
    for column in df.columns:
        df[column] = pd.to_numeric(df[column], errors='coerce')  # Convert to float, coercing errors
    
    return df

update_types(data)
print(data.dtypes)

#4- Use describe to have an overview on the data set
print("\n4\n")
print("\nDescribe: ")
data.describe()

#5- Delete the rows with missing values
print("\n5\n")
data.dropna(inplace= True)
#print("apres ", data.isna().sum())


# 6
print("\n6\n")
#print("avant \n", data.head(1))
data.loc[:,'Sub_metering_1'] = (data['Sub_metering_1'] + 1) * 0.06
#print("\naprest \n", data.head(1))

#7 
print("\n7\n")
filtered_data=data[(data.index >= '2008-12-27') & (data['Voltage'] >= 242)]
print(filtered_data.head().to_markdown())
print("Taille", len(filtered_data))

#8
print("\n8\n")
print(data.iloc[88887]) 

#9
print("\n9\n")
max_date = data['Global_active_power'].idxmax()
print("Date avec Global_active_power maximal :", max_date)

#10
print("\n10\n")
sorted_df = data.sort_values(by=['Global_active_power', 'Voltage'], ascending=[False, True])
print(sorted_df.tail().to_markdown())

#10
print("\n11\n")
daily_mean = data.groupby(data.index.date)['Global_active_power'].mean()
print("\nMoyenne journalière de Global_active_power :")
print(daily_mean)
```

```python
#Exercise 4
import pandas as pd
import numpy as np

path="data/iris.txt"
df= pd.read_csv(path, sep=",", low_memory=False)
df.drop(columns=['flower'], inplace= True)

#1

for column in df.columns:
        df[column] = pd.to_numeric(df[column], errors='coerce')

print(df.describe())
print("\n\n")

df['sepal_length'].fillna(df['sepal_length'].mean(), inplace=True)  
df['sepal_width'].fillna(df['sepal_width'].median(), inplace=True)    
df['petal_length'].fillna(0, inplace=True)                           
df['petal_width'].fillna(0, inplace=True)                         


#2

for column in df.columns:
        df[column].fillna(df[column].median(), inplace=True)

print(df.describe())

#Bonus

# Trouver une ligne spéciale
special_row = df[(df['sepal_length'] > 7) & (df['petal_length'] == 0)]
print("\nspecial rows\n", special_row.head(1))

#Trouver un sepal_length élevé avec une petal_length de 0 pourrait signaler un cas atypique ou erroné. 
#Cela pourrait indiquer des données potentiellement incorrectes, car on pourrait s'attendre à ce qu'une 
#fleur avec un grand sépale ait également des pétales de longueur non nulle.
```
